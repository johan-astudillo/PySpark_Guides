# FILL NULLS WITH DEFAULT VALUE

```
from pyspark.sql.types import IntegerType,BooleanType,DateType
df = df1.withColumn("Sales",df1.Sales.cast('integer'))
```

```
df.show()
```

## String

```
df_string = df.na.fill('CHOOSE A VALUE') # This is for string
display(df_string)
```

```
# Other way
df.na.fill('NO NAME', subset=['Name']).show()
```

## Numeric

```
df_numeric = df.na.fill(0) # This if for Numeric, Integer...
display(df_numeric)
```
